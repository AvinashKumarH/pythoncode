```python
import pandas as pd
```

```python
holidays = pd.read_csv('/content/holidays_events.csv')
```

```python
train = pd.read_csv('/content/train.csv')
```

```python
test = pd.read_csv('/content/test.csv')
```

```python
stores = pd.read_csv('/content/stores.csv')
```

```python
oil = pd.read_csv('/content/oil.csv')
```

```python
print(oil.head())
```

```python
oil.isnull().sum()
```

```python
# Handle missing values in oil prices using interpolation
oil['dcoilwtico'] = oil['dcoilwtico'].interpolate()
```

```python
# convert date columns to date time format
holidays['date'] = pd.to_datetime(holidays['date'])
print(holidays.info())
```

```python
train['date'] = pd.to_datetime(train['date'])
test['date'] = pd.to_datetime(test['date'])
oil['date'] = pd.to_datetime(oil['date'])
```

```python
# Merge datasets
train = train.merge(stores, on='store_nbr', how='left')
train = train.merge(oil, on='date', how='left')
train = train.merge(holidays, on='date', how='left')

test = test.merge(stores, on='store_nbr', how='left')
test = test.merge(oil, on='date', how='left')
test = test.merge(holidays, on='date', how='left')
```

```python
print(train.head())
```

```python
# Step 1: Time-Based Features
train['day'] = train['date'].dt.day
train['week'] = train['date'].dt.isocalendar().week
train['month'] = train['date'].dt.month
train['year'] = train['date'].dt.year
train['day_of_week'] = train['date'].dt.dayofweek  # Monday=0, Sunday
```

```python
# Identify seasonal trends (e.g., higher sales in December)
train['is_december'] = (train['month'] == 12).astype(int)
```

```python
train['is_holiday'] = train['date'].apply(lambda x: 1 if x in ['Holiday', 'Additional', 'Bridge'] else 0)
```

```python
# Flag for paydays (15th and last day of the month)
train['is_payday'] = ((train['day'] == 15) | (train['day'] == train['date'].dt.days_in_month)).astype(int)
```

```python
# Flag for earthquake impact (April 16, 2016)
train['earthquake_impact'] = (train['date'] == '2016-04-16').astype(int)
```

```python
# Step 3: Rolling Statistics
train = train.sort_values(['store_nbr', 'family', 'date'])
```

```python
train['rolling_mean_7'] = train.groupby(['store_nbr', 'family'])['sales'].transform(lambda x: x.rolling(7, min_periods=1).mean())
train['rolling_std_7'] = train.groupby(['store_nbr', 'family'])['sales'].transform(lambda x: x.rolling(7, min_periods=1).std())
```

```python
# Lagged features (previous week's sales)
train['lag_7'] = train.groupby(['store_nbr', 'family'])['sales'].shift(7)
train['lag_30'] = train.groupby(['store_nbr', 'family'])['sales'].shift(30)
```

```python
# Step 4: Store-Specific Aggregations
# Average sales per store type
store_sales = train.groupby('store_nbr')['sales'].mean().rename("avg_store_sales")
train = train.merge(store_sales, on='store_nbr', how='left')
```

```python
# Identify top-selling product families per cluster
top_products = train.groupby(['cluster', 'family'])['sales'].sum().reset_index()
top_products = top_products.sort_values(['cluster', 'sales'], ascending=[True, False])
top_products['top_product'] = top_products.groupby('cluster')['family'].transform(lambda x: x == x.iloc[0])
train = train.merge(top_products[['cluster', 'family', 'top_product']], on=['cluster', 'family'], how='left')
```

```python
test['naive_forecast'] = test['store_nbr'].shift(1)  # Last known sales as the next day's prediction
```

```python
from statsmodels.tsa.arima.model import ARIMA

arima_model = ARIMA(train['sales'], order=(5,1,0))  # (p,d,q) values tuned
arima_model_fit = arima_model.fit()
test['arima_forecast'] = arima_model_fit.forecast(steps=len(test))
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(train[['day', 'month', 'year', 'store_nbr']], train['sales'])
test['rf_forecast'] = rf.predict(test[['day', 'month', 'year', 'store_nbr']])
```

```python
from xgboost import XGBRegressor

xgb = XGBRegressor(objective='reg:squarederror', n_estimators=100)
xgb.fit(train[['day', 'month', 'year', 'store_nbr']], train['sales'])
test['xgb_forecast'] = xgb.predict(test[['day', 'month', 'year', 'store_nbr']])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

X_train = train[['sales']].values.reshape(-1, 1, 1)  # Reshape for LSTM input
y_train = train['sales'].values

model = Sequential([
    LSTM(50, activation='relu', input_shape=(1,1), return_sequences=True),
    LSTM(50, activation='relu'),
    Dense(1)
])
model.compile(optimizer='adam', loss='mse')

model.fit(X_train, y_train, epochs=10, verbose=1)
X_test = test[['sales']].values.reshape(-1, 1, 1)
test['lstm_forecast'] = model.predict(X_test)
```

```python
from prophet import Prophet

prophet_df = train[['date', 'sales']].rename(columns={'date': 'ds', 'sales': 'y'})
prophet = Prophet()
prophet.fit(prophet_df)

future = prophet.make_future_dataframe(periods=15)
forecast = prophet.predict(future)
test['prophet_forecast'] = forecast['yhat'].iloc[-len(test):].values
```

Model Evaluation: Comparing Forecasting Models
To evaluate our models, we'll use three key metrics:

Root Mean Squared Error (RMSE) → Measures the average error magnitude.

Mean Absolute Percentage Error (MAPE) → Measures the percentage error.

R-Squared Score → Measures how well the model fits the data.

Visual Inspection → Plot actual vs. predicted sales.



```python
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score
import numpy as np
import matplotlib.pyplot as plt
```

```python
def evaluate_model(y_true, y_pred, model_name):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = mean_absolute_percentage_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    print(f"📊 {model_name} Evaluation:")
    print(f"✅ RMSE: {rmse:.2f}")
    print(f"✅ MAPE: {mape:.2%}")
    print(f"✅ R² Score: {r2:.2f}")
    print("-" * 40)

    return rmse, mape, r2

# Apply the function to all models
results = {}
for model_name in ["naive_forecast", "arima_forecast", "rf_forecast", "xgb_forecast", "lstm_forecast"]:
    results[model_name] = evaluate_model(test["sales"], test[model_name], model_name)
```

```python
import pandas as pd

results_df = pd.DataFrame(results, index=["RMSE", "MAPE", "R² Score"]).T
print(results_df.sort_values("RMSE"))
```

```python
plt.figure(figsize=(14, 6))

plt.plot(test["date"], test["sales"], label="Actual Sales", color="blue")
plt.plot(test["date"], test["xgb_forecast"], label="XGBoost Prediction", color="red", linestyle="dashed")
plt.plot(test["date"], test["lstm_forecast"], label="LSTM Prediction", color="green", linestyle="dashed")

plt.title("Actual vs. Predicted Sales")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
